Digit Recoginization ML Model using tenserflow

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
mnist = tf.keras.datasets.mnist    ##loding digit dataset 
```

Spliting data into test and train

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
print('x_train shape:',x_train.shape,'x_train dtype:',x_train.dtype,'\n y_train shape:',y_train.shape,'y_train dtype:',y_train.dtype)
print('x_test shape:',x_test.shape,'x_test dtype:',x_test.dtype,'\n y_test shape:',y_test.shape,'y_test dtype:',y_test.dtype)
```

```python
print(x_train[0])
```

```python
plt.imshow(x_train[0])
plt.title(y_train[0])
plt.show()
```

Preprocessing

```python
x_train = tf.keras.utils.normalize(x_train,axis=1)
x_test = tf.keras.utils.normalize(x_test,axis=1)
x_train[0]
```

Creating Model

```python
model = tf.keras.models.Sequential() 
```

```python
model.add(tf.keras.layers.Flatten(input_shape = (28,28) ) )
model.add(tf.keras.layers.Dense(128, activation = 'relu') )
model.add(tf.keras.layers.Dropout(0.2) )
model.add(tf.keras.layers.Dense(16, activation = 'relu') )
model.add(tf.keras.layers.Dropout(0.2) )
model.add(tf.keras.layers.Dense(10, activation = 'softmax'))
```

```python
model.compile(optimizer ='adam', loss ='SparseCategoricalCrossentropy', metrics =['accuracy'])
```

```python
hist = model.fit(x_train, y_train, epochs =500, validation_split =0.25, batch_size = 1000)
```

```python
y_prob = model.predict(x_test)
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, y_pred)
print(classification_report(y_test, y_pred))
```

```python
test_loss, test_acc = model.evaluate(x_test,y_test)
print('accuracy :',test_acc)
print('loss :',test_loss)
```

```python
plt.plot(hist.history['loss'] , label = 'loss')
plt.plot(hist.history['val_loss'], label = 'val_loss')
plt.title('LOSS')
plt.legend()
plt.show()
```

```python
plt.plot(hist.history['accuracy'], label = 'accuracy')
plt.plot(hist.history['val_accuracy'], label = 'val_accuracy')
plt.title('ACCURACY')
plt.legend()
plt.show()
```
